```python
import requests
import numpy as np
import pandas as pd
import herepy
import config
import json
from datetime import datetime as dt
```

```python
residents = pd.read_csv("addresses_simulated_incomes.csv")
car_mask = residents['simulated_commute_car']==1
residents_car = residents[car_mask]
residents_transit = residents[car_mask]
#Need to add store and vacant lot locations into a single df. Idk where current store info is--can you add that and the
#vacant lot data here?
lots = pd.read_csv("vacant_lots_centroid.csv")
stores = pd.read_csv("supermarkets.csv")
```

```python
#limited rows for testing. Change to residents_car.loc[:, [..]].values instead of the slices to use whole datasets
start_car_coords = residents_car.loc[0:5,["LATITUDE", "LONGITUDE"]].values
start_transit_coords = residents_transit.loc[0:5,["LATITUDE", "LONGITUDE"]].values
end_coords_lots = lots.loc[0:5,["Y", "X"]].values
end_coords_stores = stores.loc[0:5,["Latitude","Longitude"]].values
end_coords = np.concatenate((end_coords_lots,end_coords_stores),axis=0)

routingApi = herepy.RoutingApi(config.api_key)
publicTransitApi = herepy.PublicTransitApi(config.api_key)
```

```python
#use car_api_start and car_api_end to time how long the API takes to run
car_api_start = dt.now()
response_car = routingApi.matrix(
    start_waypoints=start_car_coords,
    destination_waypoints=end_coords,
    departure='2020-07-04T17:00:00+02',
    summary_attributes = ['traveltime'],
    modes=[herepy.RouteMode.fastest, herepy.RouteMode.car])
car_api_end = dt.now()
car_api_end-car_api_start
```

```python
#The routingApi.matrix function doesn't support the publicTransport RouteMode-- will need to use the
#public_transport method below
transit_api_start = dt.now()
response_transit = publicTransitApi.calculate_route(
    departure = list(start_transit_coords),
    arrival = list(end_coords[0]),
    time = '2020-07-04T17:00:00+02')
transit_api_end = dt.now()


```
